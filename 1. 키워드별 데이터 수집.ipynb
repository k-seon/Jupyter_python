```python
# 브라우저(Chrome) 자동 제어 라이브러리
from selenium import webdriver
# 다운로드 시간 제어
import time

# 크롬 사용, 경로에 크롬 드라이버가 있어야 함.
driver = webdriver.Chrome("./chromedriver")
# 크롤링할 키워드들을 여러개 입력하기 위해 배열로 저장
keywords = [""]
for i,keyword in enumerate(keywords):
    keywords[i] = keyword + " 보안"
    
# 크롤링할 최대 페이지 num, 내용이 없으면 자동 종료. 1000으로 지정해도 됌.
max_page = 1000

for keyword in keywords:
    for i in range(max_page):
        # 묵시적 대기. 응답을 기다림.
        driver.implicitly_wait(100)
        url="http://www.riss.kr/search/Search.do?isDetailSearch=N&searchGubun=true&viewYn=OP&queryText=&strQuery=KEYWORD&exQuery=regnm%3AKCI%EB%93%B1%EC%9E%AC%E2%97%88regnm%3AKCI%EC%9A%B0%EC%88%98%EB%93%B1%EC%9E%AC%E2%97%88regnm%3AKCI%EB%93%B1%EC%9E%AC%ED%9B%84%EB%B3%B4%E2%97%88&exQueryText=%EB%93%B1%EC%9E%AC%EC%A0%95%EB%B3%B4+%5BKCI%EB%93%B1%EC%9E%AC%5D%40%40regnm%3AKCI%EB%93%B1%EC%9E%AC%E2%97%88%EB%93%B1%EC%9E%AC%EC%A0%95%EB%B3%B4+%5BKCI%EC%9A%B0%EC%88%98%EB%93%B1%EC%9E%AC%5D%40%40regnm%3AKCI%EC%9A%B0%EC%88%98%EB%93%B1%EC%9E%AC%E2%97%88%EB%93%B1%EC%9E%AC%EC%A0%95%EB%B3%B4+%5BKCI%EB%93%B1%EC%9E%AC%ED%9B%84%EB%B3%B4%5D%40%40regnm%3AKCI%EB%93%B1%EC%9E%AC%ED%9B%84%EB%B3%B4%E2%97%88&order=%2FDESC&onHanja=false&strSort=DATE&p_year1=&p_year2=&iStartCount=0&orderBy=&fsearchMethod=search&sflag=1&isFDetailSearch=N&pageNumber=&resultKeyword=&fsearchSort=&fsearchOrder=&limiterList=&limiterListText=&facetList=&facetListText=&fsearchDB=&icate=re_a_kor&colName=re_a_kor&pageScale=100&isTab=Y&regnm=&dorg_storage=&language=&language_code=&query=KEYWORD"
        url_tmp=url.replace("KEYWORD",keyword)
        url_tmp=url_tmp.replace("iStartCount=0","iStartCount="+str((i)*100))
    
        driver.get(url_tmp)
        main_window = driver.current_window_handle
    
        doc = driver.find_element_by_xpath("//div[@class='srchResultListW']/ul").text
        
        if doc=="":
            print("=========================Empty Page")
            print("Keyword : ", keyword,"Current page : ",i,"\n")
            break

        driver.find_element_by_xpath("//div[@class='resultTop1']/label[@for='allchk2']/span[@class='checkBox']").click()
        driver.execute_script("window.scrollTo(0,0)")
        driver.find_element_by_xpath("//div[@class='resultTop1']/ul/li/a[@href='#redirect']").click()

        
    
        popup_window = driver.window_handles[1]
        driver.switch_to_window(popup_window)
        driver.implicitly_wait(100)

        driver.find_element_by_xpath("//label[@for='radio-3']").click()
        driver.find_element_by_xpath("//label[@for='radio-8']").click()
        driver.find_element_by_xpath("//label[@for='radio-12']").click()
        driver.find_element_by_xpath("//div[@class='btnBunch']/a[@href='#redirect']").click()
        time.sleep(9)
    
        driver.switch_to_window(popup_window)
        driver.close()
        driver.switch_to_window(main_window)
        
driver.close()
```
